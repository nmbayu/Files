```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
```

```python
# Membuat tanggal acak dalam rentang tertentu
def random_dates(start, end, n=10):
    start_u = start.value//10**9
    end_u = end.value//10**9
    return pd.to_datetime(np.random.randint(start_u, end_u, n), unit='s')
```

```python
# Menentukan rentang tanggal
start = pd.to_datetime('2022-01-01')
end = pd.to_datetime('2022-12-31')
```

```python
# Membuat DataFrame
data = pd.DataFrame(index=random_dates(start, end, 1000))
data = data.sort_index()
```

```python
# Membuat kolom 'pasien_masuk' dan 'pasien_keluar' dengan nilai acak
data['pasien_masuk'] = [random.randint(10, 50) for _ in range(len(data))]
data['pasien_keluar'] = [random.randint(5, data.loc[date, 'pasien_masuk']) for date in data.index]
```

```python
# Menampilkan data
print(data)
```

```python
import matplotlib.pyplot as plt
```

```python
# Membuat plot
plt.figure(figsize=(10, 6))
```

```python
# Plot jumlah pasien yang masuk
plt.plot(data.index, data['pasien_masuk'], label='Pasien Masuk')
```

```python
# Plot jumlah pasien yang keluar
plt.plot(data.index, data['pasien_keluar'], label='Pasien Keluar')
```

```python
# Menambahkan judul dan label
plt.title('Jumlah Pasien Masuk dan Keluar')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Pasien')
plt.legend()
```

```python
# Menampilkan plot
plt.show()
```

```python
# Menghitung total pasien masuk dan keluar
total_masuk = data['pasien_masuk'].sum()
total_keluar = data['pasien_keluar'].sum()
```

```python
# Membuat plot
plt.figure(figsize=(6, 6))
```

```python
# Data untuk pie chart
sizes = [total_masuk, total_keluar]
labels = ['Pasien Masuk', 'Pasien Keluar']
```

```python
# Membuat pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
```

```python
# Menambahkan judul
plt.title('Persentase Pasien Masuk dan Keluar')
```

```python
# Menampilkan plot
plt.show()
```
